## The DSPy Paradigm

### Traditional vs DSPy Approach

```python
import dspy
from dotenv import load_dotenv

# Load environment variables
load_dotenv()

# Configure DSPy
lm = dspy.OpenAI(model="gpt-3.5-turbo")
dspy.settings.configure(lm=lm)
```

### Traditional (Imperative) Approach

```python
# Traditional approach - manually crafting prompts
traditional_prompt = """
You are a helpful assistant. 
Answer the following question: {question}
"""

# This requires manual prompt engineering
print("Traditional approach requires manual prompt crafting")
```

### DSPy (Declarative) Approach

```python
# DSPy approach - declare what you want
class QuestionAnswering(dspy.Signature):
    """Answer questions with short, factual responses."""
    question = dspy.InputField(desc="the question to answer")
    answer = dspy.OutputField(desc="the answer to the question")

# DSPy handles the implementation
qa = dspy.ChainOfThought(QuestionAnswering)

# Test it
result = qa(question="What is AI?")
print(f"Answer: {result.answer}")
```

## Core Concepts

### 1. Signatures
Signatures define the input/output interface.

```python
class TextSummary(dspy.Signature):
    """Summarize the given text concisely."""
    text = dspy.InputField(desc="text to summarize")
    summary = dspy.OutputField(desc="concise summary")

print("Signature defines the interface")
```

### 2. Modules
Modules are building blocks for DSPy programs.

```python
# Basic module
summarizer = dspy.Predict(TextSummary)

# Chain of Thought module
cot_summarizer = dspy.ChainOfThought(TextSummary)

print("Modules process data according to signatures")
```

## Exercise: Build Your First DSPy Program

```python
# Your turn! Create a signature for sentiment analysis

class SentimentAnalysis(dspy.Signature):
    """Analyze the sentiment of the given text."""
    text = dspy.InputField(desc="text to analyze")
    sentiment = dspy.OutputField(desc="positive, negative, or neutral")
    confidence = dspy.OutputField(desc="confidence score from 0 to 1")

# Create the program
sentiment_analyzer = dspy.Predict(SentimentAnalysis)

# Test it
result = sentiment_analyzer(text="I love DSPy! It's amazing for building AI systems.")
print(f"Sentiment: {result.sentiment}")
print(f"Confidence: {result.confidence}")
```

## Exercise: Chain of Thought

Use ChainOfThought for complex reasoning tasks.

```python
class MathSolver(dspy.Signature):
    """Solve math problems step by step."""
    problem = dspy.InputField(desc="math problem to solve")
    reasoning = dspy.OutputField(desc="step-by-step reasoning")
    answer = dspy.OutputField(desc="final answer")

math_solver = dspy.ChainOfThought(MathSolver)

result = math_solver(problem="If a train travels at 60 mph for 2.5 hours, how far does it travel?")
print(f"Reasoning: {result.reasoning}")
print(f"Answer: {result.answer}")
```
